```python
def factorial(x):
    fact = 1
    for i in range(1,x+1):
        fact = fact*i
    return(fact)

```

```python
def factorial(x):
    if x==1:
        return 1
    else:
        return x*factorial(x-1)
```

```python
def fibonacci(x):
    if (x==1):
        return 0
    elif (x==2):
        return 1
    else:
        return (fibonacci(x-1)+fibonacci(x-2))
    
```
